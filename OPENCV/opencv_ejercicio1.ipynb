```python
#Carga de libreria 
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Recursos/ML.png')
plt.imshow(img)
```

```python
#
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(imgGray)
```

```python
imgHSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV ) 
plt.imshow(imgHSV)
```

```python
imgBlur = cv2.GaussianBlur(imgGray,(3,3),0)
plt.imshow(imgBlur)
```

```python
imgCanny = cv2.Canny(img, 21,200)
plt.imshow(imgCanny)
```

```python
kernelDilatacion = np.ones((3,3),np.uint8)
imgDilatacion = cv2.dilate(imgCanny,kernelDilatacion,iterations=1)
plt.imshow(imgDilatacion)
```

```python
kernelErosion = np.ones((1,1),np.uint8)
imgErosion = cv2.dilate(imgDilatacion,kernelErosion,iterations=2)
plt.imshow(imgErosion)
```

```python
kernelApertura = np.ones((1,1),np.uint8)
imgApertura = cv2.morphologyEx(imgCanny, cv2.MORPH_OPEN, kernelApertura,iterations=1)
plt.imshow(imgApertura)
```

```python
kernelCierre = np.ones((1,1),np.uint8)
imgCierre = cv2.morphologyEx(imgCanny, cv2.MORPH_CLOSE, kernelCierre,iterations=1)
plt.imshow(imgCierre)
```
